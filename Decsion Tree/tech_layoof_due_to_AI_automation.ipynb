# Predict layoffs_count
# Using DecisionTreeRegressor

````python
y_reg = df['layoffs_count']
X_reg = df.drop(columns=['layoffs_count'])
````

````python
X_reg = pd.get_dummies(
    X_reg,
    drop_first=True
)
````

````python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.2,
    random_state=42
)
````

````python
from sklearn.tree import DecisionTreeRegressor
reg_model = DecisionTreeRegressor(

    criterion='squared_error',

    max_depth=8,
    min_samples_split=10,
    min_samples_leaf=5,

    random_state=42
)
reg_model.fit(
    X_train_reg,
    y_train_reg
)
y_reg_pred = reg_model.predict(X_test_reg)
````

````python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
mae = mean_absolute_error(
    y_test_reg,
    y_reg_pred
)

mse = mean_squared_error(
    y_test_reg,
    y_reg_pred
)

rmse = np.sqrt(mse)

r2 = r2_score(
    y_test_reg,
    y_reg_pred
)
print("MAE :", mae)

print("MSE :", mse)

print("RMSE:", rmse)

print("R2 Score:", r2)
````

````python
feature_importance = pd.DataFrame({

    'Feature': X_reg.columns,
    'Importance': reg_model.feature_importances_

})

feature_importance = feature_importance.sort_values(
    by='Importance',
    ascending=False
)


print("\nTop 10 Important Features:")
print(feature_importance.head(10))
````

````python
plt.figure(figsize=(10,6))

sns.barplot(

    x=feature_importance.head(10)['Importance'],
    y=feature_importance.head(10)['Feature']

)

plt.title("Top 10 Feature Importances")
plt.xlabel("Importance Score")
plt.ylabel("Features")

plt.show()
````
