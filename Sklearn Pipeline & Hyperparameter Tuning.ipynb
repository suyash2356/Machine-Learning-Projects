```python
#Example 3: More Complex Pipeline
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
#For numeric processing pipeline
numeric_processor = Pipeline(
    steps = [
        ("impute_mean", SimpleImputer(missing_values=np.nan,strategy="mean")),
        ("Sclar", StandardScaler())
    ]
)
```

```python
numeric_processor
```

```python
#For categorical processing pipeline
from sklearn.preprocessing import OneHotEncoder
categorical_processor = Pipeline(
    steps = [
        ("impute_constant", SimpleImputer(fill_value="missing",strategy="constant")),
        ("OneHot",OneHotEncoder(handle_unknown="ignore"))
    ]
)
```

```python
categorical_processor
```

```python
#COmbine Processing techniques
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    [("Categorical",categorical_processor,["gender","City"]),
     ("numerical",numeric_processor,["age","height"])]
)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(preprocessor, LogisticRegression())
```

```python
pipe
```

```python
#Hyperparameter Tuning Using Machine Learning
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
x = df.iloc[:,1:]
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
##Pipeline
numeric_processor = Pipeline(
    steps=[
        ("impute_mean", SimpleImputer(missing_values=np.nan, strategy="mean")),
        ("Scalar",StandardScaler()),
    ]
)
```

```python
categorical_processor = Pipeline(
    steps = [
        (
            "imputation_constant", 
            SimpleImputer(fill_value="missing", strategy="constant"),
        ),
        ("onehot",OneHotEncoder(handle_unknown="ignore")),
    ]
)
```

```python
preprocessor = ColumnTransformer(
    [
        ("categorical",categorical_processor,["sex","smoker","day","time"]),
        ("numerical",numeric_processor,["tip","size"]),
    ]
)
```

```python
pipe = Pipeline(
    steps = [
        ("preprocessor",preprocessor),("regressor",RandomForestRegressor())
    ]
)
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.predict(x_test)
```

```python
#Hyperparameter Tuning
param_grid = {
    "regressor__n_estimators":[200,500],
    "regressor__max_features":["auto","sqrt","log2"],
    "regressor__max_depth":[4,5,6,7,8]
}
```

```python
grid_search = GridSearchCV(pipe, param_grid=param_grid, n_jobs=1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
pipe = Pipeline(
    steps = [
        ("preprocessor",preprocessor),("regressor",RandomForestRegressor(max_depth=5,max_features='sqrt',n_estimators=500))
    ]
)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
#Evaluate the model
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
