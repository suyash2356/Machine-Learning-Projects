```python
#Feature Engineering
symptom_cols = [
    "Fever",
    "Tiredness",
    "Dry-Cough",
    "Difficulty-in-Breathing",
    "Sore-Throat",
    "Pains",
    "Nasal-Congestion",
    "Runny-Nose",
    "Diarrhea"
]

df["Symptoms_Count"] = df[symptom_cols].sum(axis=1)
```

```python
resp_cols = [
    "Dry-Cough",
    "Difficulty-in-Breathing",
    "Sore-Throat"
]

df["Respiratory_Score"] = df[resp_cols].sum(axis=1)
```

```python
nasal_cols = [
    "Nasal-Congestion",
    "Runny-Nose"
]

df["Nasal_Score"] = df[nasal_cols].sum(axis=1)
```

```python
df["Digestive_Score"] = df["Diarrhea"]
```

```python
df["Exposure_Risk"] = (
    2 * df["Contact_Yes"] +
    1 * df["Contact_Dont-Know"] +
    0 * df["Contact_No"]
)
```

```python
df["High_Risk_Age"] = df["Age_60+"]
```

```python
df["Severity_Score"] = (
    0 * df["Severity_None"] +
    1 * df["Severity_Mild"] +
    2 * df["Severity_Moderate"] +
    3 * df["Severity_Severe"]
)
```

```python
df["Risk_Score"] = (
    df["Symptoms_Count"] +
    df["Respiratory_Score"] +
    df["Exposure_Risk"] +
    df["High_Risk_Age"]
)
```

```python
new_features = [
    "Symptoms_Count",
    "Respiratory_Score",
    "Nasal_Score",
    "Digestive_Score",
    "Exposure_Risk",
    "High_Risk_Age",
    "Severity_Score",
    "Risk_Score"
]

df[new_features].describe()
```

```python
features = [
    "Fever",
    "Tiredness",
    "Dry-Cough",
    "Difficulty-in-Breathing",
    "Sore-Throat",
    "Pains",
    "Nasal-Congestion",
    "Runny-Nose",
    "Diarrhea",

    "Gender_Female",
    "Gender_Male",
    "Gender_Transgender",

    "Symptoms_Count",
    "Respiratory_Score",
    "Nasal_Score",
    "Digestive_Score",
    "Exposure_Risk",
    "High_Risk_Age",
    "Severity_Score",
    "Risk_Score"
]

X = df[features]

print("Shape:", X.shape)
X.head()
```

```python
# X is your selected feature set

from sklearn.ensemble import IsolationForest

iso_no_scale = IsolationForest(
    contamination=0.03,
    random_state=42
)

pred_no_scale = iso_no_scale.fit_predict(X)

scores_no_scale = iso_no_scale.decision_function(X)

print("Normal:", (pred_no_scale == 1).sum())
print("Anomaly:", (pred_no_scale == -1).sum())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
iso_scaled = IsolationForest(
    contamination=0.03,
    random_state=42
)

pred_scaled = iso_scaled.fit_predict(X_scaled)

scores_scaled = iso_scaled.decision_function(X_scaled)

df["IsolationForest_Anomaly"] = (pred_scaled == -1).astype(int)

print("Normal:", (pred_scaled == 1).sum())
print("Anomaly:", (pred_scaled == -1).sum())
```

```python
print("Without Scaling")
print((pred_no_scale == -1).sum())

print("\nWith Scaling")
print((pred_scaled == -1).sum())
```

```python
plt.figure(figsize=(10,5))

plt.hist(scores_no_scale,
         bins=50,
         alpha=0.5,
         label="Without Scaling")

plt.hist(scores_scaled,
         bins=50,
         alpha=0.5,
         label="With Scaling")

plt.legend()
plt.title("Isolation Forest Score Distribution")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca_no = pca.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca_no = pca.fit_transform(X)
```

```python
X_pca_scaled = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(
    X_pca_no[:,0],
    X_pca_no[:,1],
    c=(pred_no_scale == -1),
    alpha=0.5
)
plt.title("Without Scaling")

plt.subplot(1,2,2)
plt.scatter(
    X_pca_scaled[:,0],
    X_pca_scaled[:,1],
    c=(pred_scaled == -1),
    alpha=0.5
)
plt.title("With Scaling")

plt.show()
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

clusters = kmeans.fit_predict(X_scaled)

distances = np.min(
    cdist(X_scaled, kmeans.cluster_centers_),
    axis=1
)

threshold = np.percentile(distances, 97)

kmeans_anomaly = distances > threshold

df["KMeans_Anomaly"] = kmeans_anomaly.astype(int)

print("Anomalies:", kmeans_anomaly.sum())
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_pca_scaled[:,0],
    X_pca_scaled[:,1],
    c=df["KMeans_Anomaly"],
    alpha=0.5
)

plt.title("K-Means Anomalies")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=1.5,
    min_samples=10
)

db_labels = dbscan.fit_predict(X_scaled)

df["DBSCAN_Anomaly"] = (
    db_labels == -1
).astype(int)

print(
    "Anomalies:",
    (db_labels == -1).sum()
)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_pca_scaled[:,0],
    X_pca_scaled[:,1],
    c=df["DBSCAN_Anomaly"],
    alpha=0.5
)

plt.title("DBSCAN Anomalies")
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=0.03
)

lof_pred = lof.fit_predict(X_scaled)

df["LOF_Anomaly"] = (
    lof_pred == -1
).astype(int)

print(
    "Anomalies:",
    (lof_pred == -1).sum()
)
```

```python
from sklearn.svm import OneClassSVM

svm = OneClassSVM(
    nu=0.03,
    kernel="rbf",
    gamma="scale"
)

svm_pred = svm.fit_predict(X_scaled)

df["SVM_Anomaly"] = (
    svm_pred == -1
).astype(int)

print(
    "Anomalies:",
    (svm_pred == -1).sum()
)
```

```python
from scipy.stats import multivariate_normal

mean = np.mean(X_scaled, axis=0)

cov = np.cov(
    X_scaled,
    rowvar=False
)

gaussian = multivariate_normal(
    mean=mean,
    cov=cov,
    allow_singular=True
)

probs = gaussian.pdf(X_scaled)

threshold = np.percentile(
    probs,
    3
)

gaussian_anomaly = probs < threshold

df["Gaussian_Anomaly"] = (
    gaussian_anomaly.astype(int)
)

print(
    "Anomalies:",
    gaussian_anomaly.sum()
)
```

# COVID-19 Anomaly Detection Using Unsupervised Machine Learning

## Dataset Size

* Total Records: 316,800
* Total Features: 27 original features
* Data Type: Cleaned and One-Hot Encoded COVID-19 Dataset
* Learning Type: Unsupervised Learning

## Steps Performed

### 1. Data Understanding

* Explored dataset structure, feature types, and distributions.
* Identified symptom, age, gender, contact history, and severity features.

### 2. Exploratory Data Analysis (EDA)

* Analyzed symptom frequency.
* Studied severity, age group, and contact history distributions.
* Generated symptom correlation heatmap.
* Created Symptoms_Count feature and analyzed its distribution.

### 3. Feature Engineering

Created new features:

* Symptoms_Count
* Respiratory_Score
* Nasal_Score
* Digestive_Score
* Exposure_Risk
* High_Risk_Age
* Severity_Score
* Risk_Score

### 4. Data Preprocessing

* Selected relevant features.
* Applied StandardScaler.
* Performed PCA for 2D visualization and anomaly analysis.

### 5. Anomaly Detection Models

Implemented and compared:

* Isolation Forest
* K-Means
* DBSCAN
* Local Outlier Factor (LOF)
* One-Class SVM
* Gaussian Anomaly Detection

## Key Analysis

* Dataset was highly balanced across age groups, severity levels, and contact history.
* Respiratory symptoms showed moderate positive correlations.
* Most patients had 3–5 symptoms.
* Unusual symptom and severity combinations were identified as potential anomalies.

## Model Comparison

| Algorithm        | Anomalies Detected |
| ---------------- | ------------------ |
| Isolation Forest | 9,500              |
| K-Means          | 9,360              |
| DBSCAN           | 0                  |
| LOF              | 8,520              |
| One-Class SVM    | 17,620             |
| Gaussian         | 9,480              |

## Final Model

**Isolation Forest** was selected as the final model because:

* Detected anomalies consistently.
* Worked well on high-dimensional data.
* Was robust to feature scaling.
* Produced results similar to Gaussian and K-Means while being more scalable and interpretable for anomaly detection.

## Conclusion

This project demonstrated how multiple unsupervised learning algorithms can be used to identify unusual patient patterns in COVID-19 symptom data. Feature engineering significantly improved data representation, and Isolation Forest provided the most balanced and reliable anomaly detection performance among all evaluated models.
